Cleaned & preprocessed daily weather data. Outputted file ('daily_precip_snow_2000_to_2023_some_missing_data.csv') daily between-station measurement means. 

** Data sourced from https://ncei.noaa.gov/cdo-web/orders?email=jensenl7@msu.edu&id=3456458 and https://www.ncei.noaa.gov/cdo-web/orders?email=jensenl7@msu.edu&id=3456473

```python
import pandas as pd
from datetime import datetime
import plotly.express as px
from sklearn import preprocessing
import numpy as np
from epiweeks import Week
```

```python
df = pd.read_csv('daily_weather.csv')
```

```python
df = df.drop(columns=['AWND','TAVG','TSUN'])
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
for col in ['PRCP','SNOW']:
    df[col] = df[col].apply(lambda x: 0.0 if pd.isna(x)==True else 0.0 if float(x)<0 else float(x))
```

```python
df.groupby('DATE')['PRCP'].mean()
```

```python
data = pd.DataFrame()
```

```python
data['Precip'] = df.groupby('DATE')['PRCP'].mean()
```

```python
data['Snow'] = df.groupby('DATE')['SNOW'].mean()
```
