## Short notebook for ILI pivot table

```python
import pandas as pd
```

```python
dataset = pd.read_csv('dataset by week (median-filled) .csv')
```

```python
dataset.head()
```

```python
from datetime import datetime
import epiweeks
```

```python
df = dataset[['epiweek','cases']]
```

```python
df.head()
```

```python
df['year'] = df['epiweek'].apply(lambda x: str(x)[:4])
df['week'] = df['epiweek'].apply(lambda x: str(x)[4:])
```

```python
df = df.drop(columns=['epiweek']).set_index('year')
```

```python
df.head()
```

```python
df = pd.pivot(df, columns='week', values='cases')
```

```python
df.head()
```

```python
df.to_csv('pivoted week cases .csv')
```
